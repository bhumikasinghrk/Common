```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('allyears.zip',compression='zip')
```

```python
df.head()
```

```python
years=df.set_index(['sex','name','year']).sort_index()
years
```

```python
years.loc[('F','Emma')]
```

```python
plt.style.use('seaborn')
plt.plot(years.loc[('F','Emma')])
plt.title("Number of girls named Emma")
plt.xlabel("Year")
plt.ylabel("Number")
```

```python
plt.style.use('seaborn')
plt.plot(years.loc[('F','Emma')] / years.groupby('year').sum())
plt.title("Percentage of girls named Emma in a year")
plt.xlabel("Year")
plt.ylabel("Number")
```

```python
def plotname(sex,name):
    plt.style.use('seaborn')
    
    data=years.loc[sex,name]
    plt.plot(data.index,data.values)
    #plt.title(f'number of %s named %s',(sex,name))
    plt.xlabel('Year')
    plt.ylabel('Number')
    
```

```python
def compareNames(sex,names):
    plt.figure(figsize=(20,5))
    for name in names:
        
        plotname(sex,name)
    plt.legend(names)
```

```python
compareNames('F',['Emma','Isabella','Anna','Aabha'])
```
