```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as pp
%matplotlib inline
```

```python
pd.options.display.max_rows=10
```

```python
years=pd.read_csv('allyears.zip',compression='zip')
```

```python
years
```

```python
year_index=years.set_index(['sex','year'])
```

```python
year_index
```

```python
year_index.loc['F',2018].sort_values('number',ascending=False).head(10)
```

```python
year_index.loc['M',2018].sort_values('number',ascending=False).head(10).name
```

```python
def getnames(sex,year):
    return (year_index.loc[sex,year].sort_values('number',ascending=False).head(10).reset_index().name)
    
```

```python
getnames('F',2018)
```

```python
top10malenames=pd.DataFrame({year:getnames('F',year) for year in range(2010,2019)})
top10malenames
```

```python
top10femalenames=pd.DataFrame({year:getnames('F',year) for year in range(2010,2019)})
top10femalenames
```

```python
def plotname(sex,name):
    data=years.query('sex==@sex and name==@name')
    
    pp.plot(data.year,data.number,label=name)
    pp.legend()
```

```python
plotname('F','Emma')
```

```python
pp.figure(figsize=(20,5))

for name in getnames('F',2018):
    plotname('F',name)
```

```python
pp.figure(figsize=(20,5))
for name in getnames('M',2018):
    plotname('M',name)
    
```
